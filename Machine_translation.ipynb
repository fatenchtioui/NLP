```python
import gradio as gr
from transformers import pipeline
```

```python
translation_pipeline = pipeline('translation_en_to_de')
```

```python
results =translation_pipeline('I love ice cream')
```

```python
results[0]['translation_text']
```

```python
def translation_transformers(from_text):
  results = translation_pipeline(from_text)
  return results[0]['translation_text']
```

```python
translation_transformers('My name is Faten')
```

```python
import gradio as gr

interface = gr.Interface(
    fn=translation_transformers,
    inputs=gr.Textbox(lines=2, placeholder='Text to translate'),
    outputs='text'
)
```

```python
interface.launch()
```
